```python
import yfinance as yf
```

```python
# ดึงหุ้น โดยใช้คำสั่ง Ticker
tesla =yf.Ticker('tsla')
```

```python
tesla
```

```python
#ดึงหุ้นของตลาด set.or.th
interval = '1mo' #1วัน = 1d , 1 เดือน = 1mo
start = '2022-1-1'
end = '2023-10-17'
```

```python
tesla_price = tesla.history(interval=interval,start=start,end=end)
```

```python
tesla_price
```

```python
tesla_price['Close'].plot()
```
